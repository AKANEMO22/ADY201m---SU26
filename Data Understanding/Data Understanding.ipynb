```python
import pandas as pd
import numpy as np
df = pd.read_csv("Bangladesh_database_Final_Merged.csv")

df.head()
```

```python
print("KIỂU DỮ LIỆU TỪNG CỘT")
print(df.dtypes)
print()
print("THỐNG KÊ MÔ TẢ (CỘT SỐ)")
df.describe()
```

```python
print("THỐNG KÊ CỘT CATEGORICAL")
cat_cols = ['District', 'Dominant_Soil_Texture', 'AP Ratio',
            'Growth', 'Crop Name', 'Transplant', 'Season']
 
for col in cat_cols:
    print(f" [{col}] — {df[col].nunique()} giá trị duy nhất")
    print(df[col].value_counts().head(5).to_string())
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
 
missing_df = pd.DataFrame({
    'Số lượng NaN': missing,
    'Tỷ lệ (%)': missing_pct
})
 
print("=== KIỂM TRA GIÁ TRỊ NULL ===")
print(missing_df[missing_df['Số lượng NaN'] > 0])
 
if missing.sum() == 0:
    print("\n KHÔNG có giá trị null thực sự trong dataset")
```

```python
n_dup = df.duplicated().sum()
print(f"Số hàng trùng lặp: {n_dup}")
 
if n_dup > 0:
    print("\\n--- Các hàng bị trùng ---")
    print(df[df.duplicated(keep=False)])
    df = df.drop_duplicates()
    print(f"\n Đã xoá xong. Kích thước mới: {df.shape}")
else:
    print(" Không có hàng nào trùng lặp!")
```

```python
print(f"Kiểu dữ liệu: {df['AP Ratio'].dtype}")  

 
failed_mask = pd.to_numeric(df['AP Ratio'], errors='coerce').isnull()
print(f"Số giá trị không convert được: {failed_mask.sum()}")
print("Các giá trị lỗi:")
print(df[failed_mask]['AP Ratio'].value_counts())
#Vấn đề: Cột `AP Ratio` đang là kiểu `str` và có 416 giá trị `#DIV/0!` (lỗi chia 0 từ Excel)
#xử lý lỗi 
df['AP Ratio'] = df['AP Ratio'].replace('#DIV/0!', np.nan) # Bước 1: thay lỗi div bằng NaN 
 
# Bước 2: Convert sang float
df['AP Ratio'] = pd.to_numeric(df['AP Ratio'], errors='coerce')
 
# Bước 3: Điền NaN bằng giá trị trung vị (median) — ít bị ảnh hưởng bởi outlier
median_val = df['AP Ratio'].median()
df['AP Ratio'] = df['AP Ratio'].fillna(median_val)
 
print("=== SAU KHI SỬA ===")
print(f"Kiểu dữ liệu mới : {df['AP Ratio'].dtype}")
print(f"Số NaN còn lại   : {df['AP Ratio'].isnull().sum()}")
print(f"Đã điền bằng median = {median_val:.4f}")
print(f"\n Thống kê mô tả:")
print(df['AP Ratio'].describe())
```

```python
print(" TRƯỚC KHI SỬA ")
print("Tên cột có khoảng trắng:")
for col in df.columns:
    if ' ' in col:
        print(f"    '{col}'")
 #Vấn đề: `AP Ratio`, `Avg Humidity`, `Crop Name` có dấu cách _ gây lỗi khi dùng `df.column_name
#xử lý
# ── Xử lý lỗi ──
df.columns = df.columns.str.strip().str.replace(' ', '_')
 
print("=== SAU KHI SỬA ===")
print("Tất cả tên cột:")
print(list(df.columns))
```

```python
print(df['Dominant_Soil_Texture'].value_counts())
#Vấn đề: Tên loại đất chứa chú thích tiếng Việt trong ngoặc
#xử lý 
import re
df['Dominant_Soil_Texture'] = df['Dominant_Soil_Texture'].apply(
    lambda x: re.sub(r'\\s*\\(.*?\\)', '', str(x)).strip()
)
 
print(" SAU KHI SỬA ")
print(df['Dominant_Soil_Texture'].value_counts())
print("\n Đã chuẩn hoá tên loại đất!")
```

```python
# Tìm giá trị có 2 khoảng trắng liên tiếp
double_space = df['Growth'].str.contains('  ', na=False)
print("Giá trị có 2+ khoảng trắng:", df[double_space]['Growth'].value_counts().to_dict())
 
# Tìm giá trị viết thường (ví dụ 'july' thay vì 'July')
bad_case = df['Growth'].str.contains('[a-z]o [a-z]', na=False)  # "to july"
print("\n Giá trị viết thường bất thường:")
print(df[bad_case]['Growth'].value_counts().to_dict())
#Vấn đề: `'Jan  to  March'` (2 khoảng trắng) vs `'Dec to March'` (1 khoảng trắng), `'April to july'` (chữ j thường)
#xử lý
# Bước 1: Loại khoảng trắng thừa (collapse multiple spaces)
df['Growth'] = df['Growth'].str.split().str.join(' ')
# Bước 2: Chuẩn hoá chữ hoa/thường — Title Case
df['Growth'] = df['Growth'].str.title()
 
print(" SAU KHI SỬA ")
print(df['Growth'].value_counts())
print("\n Đã chuẩn hoá cột Growth!")
```

```python

print("Các giá trị unique:")
print(sorted(df['Transplant'].unique()))
print(f"\n Tổng: {df['Transplant'].nunique()} giá trị — nhưng thực ra chỉ có 12 tháng!")
#Vấn đề: Cùng tháng nhưng có 2 cách viết — `Aug` và `August`, `Dec` và `December`…
#xử lý 
month_map = {
    'Jan': 'January',   'Feb': 'February', 'Mar': 'March',
    'Apr': 'April',     'Jun': 'June',     'Jul': 'July',
    'Aug': 'August',    'Sep': 'September','Oct': 'October',
    'Nov': 'November',  'Dec': 'December'
}
df['Transplant'] = df['Transplant'].replace(month_map)
 

print("Các giá trị unique:")
print(sorted(df['Transplant'].unique()))
print(f"\n Đã chuẩn hoá — còn {df['Transplant'].nunique()} giá trị duy nhất!")
```

```python
n_neg = (df['NDVI_Season_Mean'] < 0).sum()
print(" TRƯỚC KHI SỬA ")
print(f"Số giá trị NDVI âm: {n_neg}")
print(df[df['NDVI_Season_Mean'] < 0]['NDVI_Season_Mean'].describe())
#Vấn đề: NDVI lý thuyết thuộc [-1, 1], nhưng giá trị âm gần 0 thường là nhiễu cảm biến → clip về 0
#xử lý
# Clip về 0 vì giá trị âm rất nhỏ (≈ -0.0017) → nhiễu đo lường
df['NDVI_Season_Mean'] = df['NDVI_Season_Mean'].clip(lower=0)
 
print(" SAU KHI SỬA ")
print(f"Số giá trị NDVI < 0 còn lại: {(df['NDVI_Season_Mean'] < 0).sum()}")
print(f"NDVI min: {df['NDVI_Season_Mean'].min():.6f}")
print("\n Đã xử lý giá trị NDVI âm!")
```

```python
# 1. Ép kiểu số an toàn cho các cột đo lường khí hậu
numeric_fix_cols = ['Avg Temp', 'Avg Humidity', 'Max Temp', 'Min Temp', 'Rainfall', 'Wind_Mean', 'pH']
for col in numeric_fix_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].fillna(df[col].mean())

# 2. Hiệu chỉnh hệ số tỷ lệ cho FPAR
if 'FPAR' in df.columns:
    df['FPAR'] = df['FPAR'] * 0.1

# 3. Chuyển đổi LST từ Kelvin sang độ C
if 'LST_Kelvin' in df.columns:
    df['LST_C'] = df['LST_Kelvin'] - 273.15
    df = df.drop(columns=['LST_Kelvin'])

# 4. Chuẩn hóa tổng tỷ lệ thành phần đất (Clay + Sand + Silt = 100%)
soil_parts = ['Clay', 'Sand', 'Silt']
if all(col in df.columns for col in soil_parts):
    soil_sum = df[soil_parts].sum(axis=1)
    df[soil_parts] = df[soil_parts].div(soil_sum, axis=0) * 100

# 5. Đảo ngược lỗi hoán đổi cột độ ẩm nếu có
if 'Max Relative Humidity' in df.columns and 'Min Relative Humidity' in df.columns:
    # Đảm bảo Max luôn lớn hơn hoặc bằng Min
    mask = df['Max Relative Humidity'] < df['Min Relative Humidity']
    df.loc[mask, ['Min Relative Humidity', 'Max Relative Humidity']] = df.loc[mask, ['Max Relative Humidity', 'Min Relative Humidity']].values

print("Hoàn thành hiệu chỉnh logic và chỉ số vật lý.")
```

```python
#Phát hiện và xử lý Outlier (IQR Method)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
 
print(" PHÁT HIỆN OUTLIER ")
print(f"{'Cột':<25} {'Số outlier':>12} {'Tỷ lệ (%)':>12}")
print("-" * 52)
 
outlier_cols = []
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    pct = round(n_out / len(df) * 100, 2)
    if n_out > 0:
        outlier_cols.append(col)
        print(f"{col:<25} {n_out:>12} {pct:>11}%")
 
print(f"\nTổng số cột có outlier: {len(outlier_cols)}")
#xử lý outlier

df_cleaned = df.copy()
 
for col in outlier_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_cleaned[col] = df_cleaned[col].clip(lower=lower, upper=upper)
 
print("=== SAU KHI XỬ LÝ OUTLIER ===")
# Kiểm tra lại
total_out = 0
for col in outlier_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    n = ((df_cleaned[col] < Q1 - 1.5*IQR) | (df_cleaned[col] > Q3 + 1.5*IQR)).sum()
    total_out += n
 
print(f"Tổng outlier còn lại: {total_out}")
```
