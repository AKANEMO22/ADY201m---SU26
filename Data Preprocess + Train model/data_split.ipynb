## tách file train và test

```python
import pandas as pd
import numpy as np

file_gene = 'Agri_Data_Preprocessed.csv'  
train_ratio = 0.8
random_state = 42
df = pd.read_csv(file_gene)

print(f"Tổng số dòng data: {len(df)}")
df = df.sample(frac=1, random_state=random_state).reset_index(drop=True)
split_idx = int(len(df) * train_ratio)

df_train = df.iloc[:split_idx].copy()
df_test = df.iloc[split_idx:].copy()

print(f"Kích thước train: {df_train.shape}")
print(f"Kích thước test : {df_test.shape}")

# 3. Lưu file
df_train.to_csv('Agri_Data_Train.csv', index=False)
df_test.to_csv('Agri_Data_Test.csv', index=False)

print("Đã xuất file: Agri_Data_Train.csv và Agri_Data_Test.csv")
```
