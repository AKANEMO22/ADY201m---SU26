# SMOTE + Noise (Phiên bản cải tiến)
**Các cải tiến so với bản gốc:**
1. **Log-space interpolation** cho Area & Production (skewness 7.9 → 0.65)
2. **Stratified SMOTE per Crop** — nội suy trong không gian cùng loại cây
3. **Loại derived cols khỏi SMOTE** → tái tính lại sau (Rain_Temp_Ratio, Rootzone_Surface_Diff, Moisture_Ratio, NDVI stats)
4. **Sync đầy đủ categorical** (pH_Suitability, Water_Availability_Cat, Is_Extreme_Heat, Extreme_Heat_Risk)
5. **Reproducibility** với `np.random.seed()`
6. **Rainfall clip bằng percentile** thay vì min/max tuyệt đối
7. **Giữ float** cho Area/Production thay vì ép int sớm

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler

# =============================================================
# 0. SEED & CẤU HÌNH
# =============================================================
SEED = 42
np.random.seed(SEED)

NUM_ROWS    = 20000   # Số mẫu cần sinh
K_NEIGHBORS = 5      # Số láng giềng SMOTE
NOISE_LEVEL = 0.015  # Giảm nhẹ so với bản gốc (0.02) vì log-space đã tạo diversity

# =============================================================
# 1. LOAD DỮ LIỆU
# =============================================================
print("[1/6] Đang đọc dữ liệu gốc...")
data = pd.read_csv('Agri_Data_Cleaned.csv')
print(f"     Gốc: {data.shape[0]} dòng, {data.shape[1]} cột")

# --- Định nghĩa các nhóm cột ---
CATEGORICAL_COLS = [
    'District', 'Season', 'Crop Name', 'Transplant',
    'Growth', 'Harvest', 'pH_Suitability',
    'Dominant_Soil_Texture', 'Water_Availability_Cat',
    'Extreme_Heat_Risk', 'Is_Extreme_Heat',
    'is_extreme_Heat_Stress_Days', 'is_extreme_Wind_Max'
]

# CẢI TIẾN #3: Loại derived cols ra khỏi SMOTE — tái tính sau
DERIVED_COLS = [
    'CN_Ratio', 'Rain_Temp_Ratio',
    'Rootzone_Surface_Diff', 'Moisture_Ratio',
    'NDVI_Season_Range', 'NDVI_Season_CV', 'NDVI_Season_Std'
]

# Cột gốc dùng để nội suy (không gồm Yield và derived)
BASE_NUMERIC_COLS = [
    c for c in data.columns
    if c not in CATEGORICAL_COLS + DERIVED_COLS and c != 'Yield'
]

# CẢI TIẾN #1: Cột cần log-transform trước khi nội suy
LOG_COLS = ['Area', 'Production']  # skewness ~8 → log skewness ~0.1-0.7

print(f"     Base numeric cols: {len(BASE_NUMERIC_COLS)} cột")
print(f"     Derived cols (tái tính sau): {DERIVED_COLS}")
print(f"     Log-transform cols: {LOG_COLS}")

# =============================================================
# 2. HÀM SINH DỮ LIỆU: STRATIFIED SMOTE + NOISE
# =============================================================
def generate_smote_noise_stratified(df, n_samples, k=5, noise_level=0.015, log_cols=None):
    """
    CẢI TIẾN #2: Stratified SMOTE per Crop.
    Mỗi cây trồng được nội suy trong không gian riêng của nó,
    tránh tình trạng SMOTE trộn điểm giữa Lúa và Xoài chẳng hạn.
    
    CẢI TIẾN #1: Log-space interpolation cho cột skewness cao.
    """
    if log_cols is None:
        log_cols = []

    print(f"[2/6] Stratified SMOTE (k={k}) + Noise (scale={noise_level})...")

    # --- Tính số mẫu cần sinh cho mỗi loại cây (tỷ lệ thuận với dữ liệu gốc) ---
    crop_counts = df['Crop Name'].value_counts()
    total_orig  = len(df)
    crop_target = (crop_counts / total_orig * n_samples).round().astype(int)
    # Điều chỉnh sai số làm tròn
    diff = n_samples - crop_target.sum()
    if diff != 0:
        top_crop = crop_counts.index[0]
        crop_target[top_crop] += diff

    all_synthetic = []

    for crop_name, n_crop in crop_target.items():
        if n_crop <= 0:
            continue

        crop_df = df[df['Crop Name'] == crop_name].reset_index(drop=True)
        n_orig  = len(crop_df)

        # Nếu quá ít mẫu, dùng noise-only thay SMOTE
        if n_orig < 3:
            sampled = crop_df.sample(n=n_crop, replace=True).reset_index(drop=True)
            all_synthetic.append(sampled)
            continue

        # --- Chuẩn bị dữ liệu số ---
        data_num = crop_df[BASE_NUMERIC_COLS].fillna(crop_df[BASE_NUMERIC_COLS].median())

        # CẢI TIẾN #1: Log-transform trước khi scale
        data_log = data_num.copy()
        for col in log_cols:
            if col in data_log.columns:
                data_log[col] = np.log1p(data_log[col].clip(lower=0))

        scaler = MinMaxScaler()
        data_scaled = scaler.fit_transform(data_log)

        # --- KNN trong không gian của cùng loại cây ---
        k_actual = min(k, n_orig - 1)  # Không vượt quá số mẫu có sẵn
        nbrs = NearestNeighbors(n_neighbors=k_actual + 1, algorithm='auto').fit(data_scaled)

        # --- Chọn parent ngẫu nhiên ---
        parent_idxs = np.random.choice(n_orig, n_crop, replace=True)
        parents = data_scaled[parent_idxs]

        # --- Tìm láng giềng và nội suy ---
        indices = nbrs.kneighbors(parents, return_distance=False)
        neighbor_idxs = np.array([np.random.choice(row[1:]) for row in indices])
        neighbors = data_scaled[neighbor_idxs]

        ratios = np.random.rand(n_crop, 1)
        synthetic_scaled = parents + ratios * (neighbors - parents)

        # --- Thêm Gaussian noise ---
        noise = np.random.normal(0, noise_level, synthetic_scaled.shape)
        synthetic_scaled = np.clip(synthetic_scaled + noise, 0, 1)  # Giữ trong [0,1] sau noise

        # --- Inverse transform ---
        synthetic_num = scaler.inverse_transform(synthetic_scaled)
        synthetic_df_num = pd.DataFrame(synthetic_num, columns=BASE_NUMERIC_COLS)

        # CẢI TIẾN #1: Exp-transform ngược lại
        for col in log_cols:
            if col in synthetic_df_num.columns:
                synthetic_df_num[col] = np.expm1(synthetic_df_num[col]).clip(lower=0)

        # --- Kế thừa categorical từ parent ---
        parent_cats = crop_df.iloc[parent_idxs][CATEGORICAL_COLS].reset_index(drop=True)
        synthetic_crop = pd.concat([synthetic_df_num, parent_cats], axis=1)
        all_synthetic.append(synthetic_crop)

    result = pd.concat(all_synthetic, ignore_index=True)
    print(f"     Đã sinh: {len(result)} mẫu từ {len(crop_target)} loại cây")
    return result


# =============================================================
# 3. THỰC HIỆN SINH DỮ LIỆU
# =============================================================
synthetic_data = generate_smote_noise_stratified(
    data, n_samples=NUM_ROWS, k=K_NEIGHBORS,
    noise_level=NOISE_LEVEL, log_cols=LOG_COLS
)


# =============================================================
# 4. HẬU XỬ LÝ LOGIC
# =============================================================
print("[3/6] Hậu xử lý Logic...")

# --- A. ĐỒNG BỘ CÂY TRỒNG & THỜI GIAN ---
cols_to_sync = ['Season', 'Transplant', 'Growth', 'Harvest']
synthetic_data = synthetic_data.drop(columns=cols_to_sync, errors='ignore')

sampled_rows = []
for crop in synthetic_data['Crop Name'].unique():
    idx = synthetic_data[synthetic_data['Crop Name'] == crop].index
    orig_subset = data[data['Crop Name'] == crop][cols_to_sync]
    if not orig_subset.empty:
        sampled = orig_subset.sample(n=len(idx), replace=True).set_index(idx)
        sampled_rows.append(sampled)

synthetic_data = pd.concat([synthetic_data, pd.concat(sampled_rows)], axis=1)

# --- B. RAINFALL CLIP BẰNG PERCENTILE (CẢI TIẾN #6) ---
# Dùng percentile thay min/max để tránh bị khoá cứng khi chỉ có 1-2 mẫu
geo_climate = data.groupby(['District', 'Season'])['Rainfall'].agg(
    Rain_p05=lambda x: x.quantile(0.05),
    Rain_p95=lambda x: x.quantile(0.95)
).reset_index()

# Fallback về District nếu District+Season không match
dist_climate = data.groupby('District')['Rainfall'].agg(
    Dist_p05=lambda x: x.quantile(0.05),
    Dist_p95=lambda x: x.quantile(0.95)
).reset_index()

synthetic_data = synthetic_data.merge(geo_climate, on=['District', 'Season'], how='left')
synthetic_data = synthetic_data.merge(dist_climate, on='District', how='left')

global_min = data['Rainfall'].quantile(0.01)
global_max = data['Rainfall'].quantile(0.99)
synthetic_data['Rain_p05'] = synthetic_data['Rain_p05'].fillna(synthetic_data['Dist_p05']).fillna(global_min)
synthetic_data['Rain_p95'] = synthetic_data['Rain_p95'].fillna(synthetic_data['Dist_p95']).fillna(global_max)

synthetic_data['Rainfall'] = synthetic_data['Rainfall'].clip(
    lower=synthetic_data['Rain_p05'],
    upper=synthetic_data['Rain_p95']
).round(2)
synthetic_data.drop(columns=['Rain_p05', 'Rain_p95', 'Dist_p05', 'Dist_p95'], inplace=True)

# --- C. CLIP ÂM & LOGIC NHIỆT ĐỘ ---
non_neg_cols = [
    'Rainfall', 'Soil_Moisture_mm', 'Nitrogen', 'Organic_Carbon',
    'Wind_Max', 'Wind_Mean', 'Heat_Stress_Days',
    'sm_surface', 'sm_rootzone', 'EVI', 'LAI', 'FPAR',
    'Avg_Salinity_Index', 'Bulk_Density'
]
for col in non_neg_cols:
    if col in synthetic_data.columns:
        synthetic_data[col] = synthetic_data[col].clip(lower=0)

# NDVI trong [-1, 1]
for col in [c for c in synthetic_data.columns if 'NDVI' in c]:
    synthetic_data[col] = synthetic_data[col].clip(-1.0, 1.0)

# pH trong [3.5, 9.0]
if 'pH' in synthetic_data.columns:
    synthetic_data['pH'] = synthetic_data['pH'].clip(3.5, 9.0).round(2)

# Humidity trong [0, 100]
for col in ['Min Relative Humidity', 'Avg Humidity', 'Max Relative Humidity']:
    if col in synthetic_data.columns:
        synthetic_data[col] = synthetic_data[col].clip(0, 100).round(1)

def fix_min_mean_max(df, col_min, col_mean, col_max):
    if {col_min, col_mean, col_max}.issubset(df.columns):
        mask = df[col_min] > df[col_max]
        df.loc[mask, [col_min, col_max]] = df.loc[mask, [col_max, col_min]].values
        df[col_mean] = df[col_mean].clip(lower=df[col_min], upper=df[col_max])
    return df

synthetic_data = fix_min_mean_max(synthetic_data, 'Min Temp', 'Avg Temp', 'Max Temp')
synthetic_data = fix_min_mean_max(synthetic_data, 'Min Relative Humidity', 'Avg Humidity', 'Max Relative Humidity')
synthetic_data = fix_min_mean_max(synthetic_data, 'NDVI_Season_Min', 'NDVI_Season_Mean', 'NDVI_Season_Max')
synthetic_data[['Min Temp', 'Avg Temp', 'Max Temp']] = synthetic_data[['Min Temp', 'Avg Temp', 'Max Temp']].round(1)

# Wind: Wind_Mean <= Wind_Max
if {'Wind_Mean', 'Wind_Max'}.issubset(synthetic_data.columns):
    mask_wind = synthetic_data['Wind_Mean'] > synthetic_data['Wind_Max']
    synthetic_data.loc[mask_wind, 'Wind_Mean'] = synthetic_data.loc[mask_wind, 'Wind_Max'] * 0.8

# --- D. SOIL SUM = 100% ---
soil_cols = ['Sand', 'Silt', 'Clay']
if all(c in synthetic_data.columns for c in soil_cols):
    synthetic_data[soil_cols] = synthetic_data[soil_cols].clip(lower=0)
    total = synthetic_data[soil_cols].sum(axis=1).replace(0, 100)
    for c in soil_cols:
        synthetic_data[c] = (synthetic_data[c] / total * 100).round(2)
    # Clay = phần còn lại để đảm bảo tổng đúng 100%
    synthetic_data['Clay'] = (100.00 - synthetic_data['Sand'] - synthetic_data['Silt']).round(2).clip(lower=0)

# --- E. FLAGS (CỜ RỦI RO) — ĐỒNG BỘ 2 CHIỀU ---
THRESHOLD_HEAT_DAYS = 43.0
if 'is_extreme_Heat_Stress_Days' in synthetic_data.columns:
    # Flag=1 nhưng giá trị chưa đủ → nâng lên
    m1 = (synthetic_data['is_extreme_Heat_Stress_Days'] == 1) & (synthetic_data['Heat_Stress_Days'] < THRESHOLD_HEAT_DAYS)
    synthetic_data.loc[m1, 'Heat_Stress_Days'] = np.random.uniform(THRESHOLD_HEAT_DAYS, 65, size=m1.sum())
    # Flag=0 nhưng giá trị vượt ngưỡng → hạ xuống
    m0 = (synthetic_data['is_extreme_Heat_Stress_Days'] == 0) & (synthetic_data['Heat_Stress_Days'] >= THRESHOLD_HEAT_DAYS)
    synthetic_data.loc[m0, 'Heat_Stress_Days'] = np.random.uniform(0, 42.9, size=m0.sum())

THRESHOLD_WIND = 11.0
if 'is_extreme_Wind_Max' in synthetic_data.columns:
    m1 = (synthetic_data['is_extreme_Wind_Max'] == 1) & (synthetic_data['Wind_Max'] < THRESHOLD_WIND)
    synthetic_data.loc[m1, 'Wind_Max'] = np.random.uniform(THRESHOLD_WIND, 35.0, size=m1.sum())
    m0 = (synthetic_data['is_extreme_Wind_Max'] == 0) & (synthetic_data['Wind_Max'] >= THRESHOLD_WIND)
    synthetic_data.loc[m0, 'Wind_Max'] = np.random.uniform(2.0, 10.9, size=m0.sum())

if {'Extreme_Heat_Risk', 'Is_Extreme_Heat'}.issubset(synthetic_data.columns):
    # Is_Extreme_Heat=1 không thể đi kèm Low Risk
    mask_contra = (synthetic_data['Is_Extreme_Heat'] == 1) & (synthetic_data['Extreme_Heat_Risk'] == 'Low Risk')
    synthetic_data.loc[mask_contra, 'Extreme_Heat_Risk'] = 'High Risk'
    # Low Risk → Is_Extreme_Heat phải = 0
    synthetic_data.loc[synthetic_data['Extreme_Heat_Risk'] == 'Low Risk', 'Is_Extreme_Heat'] = 0


# =============================================================
# 5. TÁI TÍNH DERIVED COLS (CẢI TIẾN #3)
# Tất cả tính từ base cols sau khi đã được clean
# =============================================================
print("[4/6] Tái tính Derived cols từ base cols...")

# Yield = Production / Area (CẢI TIẾN #7: giữ float, không ép int sớm)
synthetic_data['Area']       = synthetic_data['Area'].clip(lower=1.0)
synthetic_data['Production'] = synthetic_data['Production'].clip(lower=0.0)
synthetic_data['Yield']      = (synthetic_data['Production'] / synthetic_data['Area']).round(4)

# CN_Ratio = Organic_Carbon / Nitrogen
if 'CN_Ratio' in data.columns:
    synthetic_data['Nitrogen'] = synthetic_data['Nitrogen'].clip(lower=0.001)
    synthetic_data['CN_Ratio'] = (synthetic_data['Organic_Carbon'] / synthetic_data['Nitrogen']).round(4)

# Rain_Temp_Ratio — tính từ công thức thực tế của dữ liệu gốc
# (gốc không phải Rain/Temp, cần dùng groupby district+season)
if 'Rain_Temp_Ratio' in data.columns:
    ref = data.groupby(['District', 'Season'])['Rain_Temp_Ratio'].median().reset_index()
    ref.columns = ['District', 'Season', 'Rain_Temp_Ratio_ref']
    synthetic_data = synthetic_data.merge(ref, on=['District', 'Season'], how='left')
    # Điều chỉnh tỷ lệ theo rainfall thực tế mới
    ref_rain = data.groupby(['District', 'Season'])['Rainfall'].median().reset_index()
    ref_rain.columns = ['District', 'Season', 'Rainfall_ref']
    synthetic_data = synthetic_data.merge(ref_rain, on=['District', 'Season'], how='left')
    safe_ref = synthetic_data['Rainfall_ref'].replace(0, np.nan).fillna(1)
    synthetic_data['Rain_Temp_Ratio'] = (
        synthetic_data['Rain_Temp_Ratio_ref'] * synthetic_data['Rainfall'] / safe_ref
    ).round(4)
    synthetic_data.drop(columns=['Rain_Temp_Ratio_ref', 'Rainfall_ref'], inplace=True)

# Rootzone_Surface_Diff = sm_rootzone - sm_surface
if 'Rootzone_Surface_Diff' in data.columns:
    synthetic_data['Rootzone_Surface_Diff'] = (
        synthetic_data['sm_rootzone'] - synthetic_data['sm_surface']
    ).round(4)

# Moisture_Ratio = sm_rootzone / sm_surface
if 'Moisture_Ratio' in data.columns:
    sm_surf_safe = synthetic_data['sm_surface'].clip(lower=0.001)
    synthetic_data['Moisture_Ratio'] = (
        synthetic_data['sm_rootzone'] / sm_surf_safe
    ).round(4)

# NDVI derived
if 'NDVI_Season_Range' in data.columns:
    synthetic_data['NDVI_Season_Range'] = (
        synthetic_data['NDVI_Season_Max'] - synthetic_data['NDVI_Season_Min']
    ).clip(lower=0).round(4)

if 'NDVI_Season_Std' in data.columns:
    # Ước tính std ~ Range / 4 (quy tắc ngón tay cái)
    synthetic_data['NDVI_Season_Std'] = (synthetic_data['NDVI_Season_Range'] / 4).round(4)

if 'NDVI_Season_CV' in data.columns:
    ndvi_mean_safe = synthetic_data['NDVI_Season_Mean'].replace(0, np.nan).fillna(0.001)
    synthetic_data['NDVI_Season_CV'] = (
        synthetic_data['NDVI_Season_Std'] / ndvi_mean_safe
    ).round(4)


# =============================================================
# 6. ĐỒNG BỘ CATEGORICAL LABELS ĐẦY ĐỦ (CẢI TIẾN #4)
# =============================================================
print("[5/6] Đồng bộ Categorical labels...")

# Dominant_Soil_Texture từ Clay/Sand/Silt
if 'Dominant_Soil_Texture' in synthetic_data.columns:
    conds = [(synthetic_data['Clay'] >= 40), (synthetic_data['Sand'] >= 50), (synthetic_data['Silt'] >= 50)]
    synthetic_data['Dominant_Soil_Texture'] = np.select(conds, ['Clayey', 'Sandy', 'Silty'], default='Loamy')

# pH_Suitability từ pH (CẢI TIẾN #4 - bản gốc bỏ qua)
if 'pH_Suitability' in synthetic_data.columns:
    conds_ph = [(synthetic_data['pH'] < 5.5), (synthetic_data['pH'] > 7.5)]
    synthetic_data['pH_Suitability'] = np.select(conds_ph, ['Acidic', 'Alkaline'], default='Optimal')

# Water_Availability_Cat từ sm_rootzone (CẢI TIẾN #4)
if 'Water_Availability_Cat' in synthetic_data.columns:
    # Dùng percentile từ dữ liệu gốc làm ngưỡng
    p33 = data['sm_rootzone'].quantile(0.33)
    p66 = data['sm_rootzone'].quantile(0.66)
    conds_water = [(synthetic_data['sm_rootzone'] >= p66), (synthetic_data['sm_rootzone'] >= p33)]
    synthetic_data['Water_Availability_Cat'] = np.select(conds_water, ['High', 'Medium'], default='Low')

    # Đồng bộ với original label distribution (nếu có nhãn khác)
    orig_labels = data['Water_Availability_Cat'].unique()
    if 'Optimal' in orig_labels:  # fallback nếu nhãn gốc khác
        synthetic_data['Water_Availability_Cat'] = synthetic_data['Water_Availability_Cat'].replace(
            {'High': 'High', 'Medium': 'Optimal', 'Low': 'Low'}
        )

# =============================================================
# 7. LƯU FILE & BÁO CÁO
# =============================================================
print("[6/6] Lưu file...")

# Sắp xếp lại cột theo thứ tự gốc
original_cols = [c for c in data.columns if c in synthetic_data.columns]
extra_cols    = [c for c in synthetic_data.columns if c not in original_cols]
synthetic_data = synthetic_data[original_cols + extra_cols]

output_file = 'Agri_Data_SMOTE_Noise_v2.csv'
synthetic_data.to_csv(output_file, index=False)

print(f"\n{'='*55}")
print(f"[THÀNH CÔNG] Đã lưu: {output_file}")
print(f"{'='*55}")
print(f"Tổng số dòng    : {len(synthetic_data):,}")
print(f"Yield trung bình: {synthetic_data['Yield'].mean():.4f}  (Gốc: {data['Yield'].mean():.4f})")
print(f"Yield std       : {synthetic_data['Yield'].std():.4f}  (Gốc: {data['Yield'].std():.4f})")
print(f"Area mean       : {synthetic_data['Area'].mean():.1f}  (Gốc: {data['Area'].mean():.1f})")
print(f"Production mean : {synthetic_data['Production'].mean():.1f}  (Gốc: {data['Production'].mean():.1f})")

# Kiểm tra tính nhất quán derived cols
print(f"\n--- Kiểm tra tính nhất quán ---")
yield_check = ((synthetic_data['Production'] / synthetic_data['Area']).round(4) == synthetic_data['Yield']).mean()
print(f"Yield = Prod/Area  : {yield_check*100:.1f}% nhất quán")
if 'Rootzone_Surface_Diff' in synthetic_data.columns:
    diff_check = ((synthetic_data['sm_rootzone'] - synthetic_data['sm_surface']).round(4) == synthetic_data['Rootzone_Surface_Diff']).mean()
    print(f"Rootzone_Diff check: {diff_check*100:.1f}% nhất quán")
if 'CN_Ratio' in synthetic_data.columns:
    cn_check = ((synthetic_data['Organic_Carbon'] / synthetic_data['Nitrogen']).round(4) == synthetic_data['CN_Ratio']).mean()
    print(f"CN_Ratio check     : {cn_check*100:.1f}% nhất quán")
```
